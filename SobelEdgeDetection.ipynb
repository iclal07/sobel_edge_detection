```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sobelEdgeDetection(img_path,bright):
    img = plt.imread(img_path)
  vertical_filter = [[-1,-2,-1],[0,0,0],[1,2,1]]
  horizontal_filter= [[-1,0,1],[-2,0,2],[-1,0,1]]
  n,m,d = img.shape

  edges_img= np.zeros_like(img)
  for row in range(3,n-2):
    for col in range (3,m-2):
        loc_pix= img[row-1: row+2, col-1:col+2, 0]

        vertical_transformed_pix= vertical_filter*loc_pix
        vertical_score= vertical_transformed_pix.sum()/4

        horizontal_transformed_pix= horizontal_filter*loc_pix
        horizontal_score= horizontal_transformed_pix.sum()/4

        edge_score= (vertical_score**2 + horizontal_score**2) **0.5
        edges_img[row,col]= [edge_score]*3
  edges_img= edges_img/ edges_img.max()
  edges_img= edges_img* bright
  plt.subplot(1, 2, 1)
  plt.imshow(img)
  plt.subplot(1, 2, 2)
  plt.imshow(edges_img)
  plt.show()
```
